```python
from lyricsgenius import Genius
import time
import pandas as pd




genius = Genius('mJE15wCPoy8WL51ozZGFMZgm_HWzcZmV6distweIjCT7ut3UA2kMfey3aTpujMfu')
genius.sleep_time = 3
genius.timeout = 5

genre_list = ['rock','pop','rap','country','rb']


for genre in genre_list:
    lyric_list = []
    print("getting songs for " + genre)
    for page in range(1,3):
        print('pageNum ' + str(page))
        time.sleep(5)
        chrts = genius.charts(time_period='all_time', chart_genre=genre,page=page,per_page=50)
        for i in range(0, len(chrts['chart_items'])): 
            time.sleep(5)
            songid = chrts['chart_items'][i]['item']['id']
            print(songid)
            lyrics_found = 0
            while lyrics_found == 0:
                try: 
                    lyrics = genius.lyrics(song_id=songid)
                    lyrics_found = 1
                except:
                    print('trying again!')
                
            print(lyrics)
            lyric_list.append(lyrics)
    df = pd.DataFrame({'lyrics':lyric_list})
    df.to_csv(genre + '.csv',index=False)
    print('saved'+ genre + '.csv' )
                               
print('done')
            
    
        
```
